## Data wrangling

## Drop data entry(row) with missing values, if any

```python
#Finding the row with missing values represented as 'nan' and -1
#Removing the row i with missing values

print('Initial Length of Dataset before cleaning: ',len(Finaldata))
i=0
while i<(len(Finaldata)):
    if ('nan' in Finaldata[i]) or (-1 in Finaldata[i]) or (99 in Finaldata[i]):
        Finaldata=np.delete(Finaldata,i,axis=0)
        i-=1
    i+=1
    
print('Final length of Dataset: ',len(Finaldata))     
```

## Removing Duplicates, if any

```python
data_dup=np.unique(Finaldata,axis=None)
print('Length of Unique data: ',len(data_dup))

if len(Finaldata)==len(data_dup):
    print('No Duplicates found')
else:
    print('Duplicates found')       
```

## Data Formatting

## Applying Calculations

```python
# Creating an array with Total number of missing teeth from 32 columns [7:38]

t_count=np.zeros(Finaldata.shape,dtype='i4')

i=0
while i<(len(Finaldata)):
    j=7
    count=0
    while j<39:
        if Finaldata[i][j]==4:
            count+=1
            t_count[i]=count
        j+=1
    i+=1
print('Number of Missing teeth:',t_count[:10]) 
```

```python
# Creating a new column with average of Pocket Depth [39:44]

p_avg=np.zeros((1750,6),dtype='f4')

i=0
while i<(len(Finaldata)):
    j=39
    x=0
    while x<6 and j<45:
        p_avg[i][x]=Finaldata[i][j]
        j+=1
        x+=1       
    i+=1

# Average of 6 columns
d_avg=np.mean(p_avg,axis=1)

# Rounding off decimal places
d_avg=np.round(d_avg,0)
print('Average of Pocket Depth:\n',d_avg[:10])
```

```python
# Classification of Periodontitis
class_pd=np.zeros(Finaldata.shape,dtype='S32')

for i,val in enumerate(d_avg):
    if val<=3:
        class_pd[i]='Normal'
    elif val>=4 and val<=5:
        class_pd[i]='Mild'
    elif val>5 and val<=7:
        class_pd[i]='Moderate'
    else:
        class_pd[i]='Advanced'

print('Classification of Periodontitis:\n',class_pd[:5])   
```

```python
# Creating T-score Column
t_score=np.zeros(Finaldata.shape,dtype='f4')

i=0
while i<(len(Finaldata)):
    j=5
    while j==5:
        t_score[i]=Finaldata[i][j]-1.2
        break 
    i+=1
    
# Rounding of decimal places to 2
t_score=np.round(t_score,2)
print('T-score:\n',t_score[:5])
```

```python
# Rounding the decinal places for 'BMD' column
bmd=np.round(Finaldata['BMD'],2)
print(bmd[:5])
```

```python
# Classification of osteoporosis:
class_bmd=np.zeros(Finaldata.shape,dtype='S32')

for i,val in enumerate(t_score):
    if val>=-1:
        class_bmd[i]='Normal'
    elif val<-1 and val>-2.5:
        class_bmd[i]='Osteopenia'
    elif val<=-2.5:
        class_bmd[i]='Osteoporosis'

print('Classification of Osteoporosis:\n',class_bmd[:5])   
```

## Adding Columns

```python
# Appending the new arrays(t_count,d_avg,t_score,class_bmd) into data(Finadata)

b=np.dtype([('SEQN','i8'),('Gender','i1'),('Age','i4'),('TOA','f8'),('BMC','f8'),('ALP','i4'),('TN_1','i4'),
            ('TN_2','i4'),('TN_3','i4'),('TN_4','i4'),('TN_5','i4'),('TN_6','i4'),('TN_7','i4'),('TN_8','i4'),('TN_9','i4'),
            ('TN_10','i4'),('TN_11','i4'),('TN_12','i4'),('TN_13','i4'),('TN_14','i4'),('TN_15','i4'),('TN_16','i4'),
            ('TN_17','i4'),('TN_18','i4'),('TN_19','i4'),('TN_20','i4'),('TN_21','i4'),('TN_22','i4'),('TN_23','i4'),
            ('TN_24','i4'),('TN_25','i4'),('TN_26','i4'),('TN_27','i4'),('TN_28','i4'),('TN_29','i4'),('TN_30','i4'),
            ('TN_31','i4'),('TN_32','i4'),('P_MSF','i4'),('P_MF','i4'),('P_DF','i4'),('P_MSL','i4'),('P_ML','i4'),('P_DL','i4'),
            ('Missing_Teeth','i4'),('Pocket_Depth', 'f4'),('Diagnosis_Pocket-Depth','S32'),('BMD','f4'),('T-score','f4'),
            ('Diagnosis_T-score','S32')])

newdata=np.zeros(d_avg.shape, dtype=b)

# Copy the data from the old array to the new array
for name in a.names:
    newdata[name] = Finaldata[name]
# Adding new arrays
newdata['Missing_Teeth']=t_count
newdata['Pocket_Depth']=d_avg
newdata['Diagnosis_Pocket-Depth']=class_pd
newdata['T-score']=t_score
newdata['BMD']=bmd
newdata['Diagnosis_T-score']=class_bmd

print(newdata[:5])
```

## Drop Columns

```python
print('Initial Number of columns: ',len(newdata[0]))
columns=('TOA','BMC','TN_1','TN_2','TN_3','TN_4','TN_5','TN_6','TN_7','TN_8','TN_9','TN_10','TN_11','TN_12','TN_13','TN_14',
         'TN_15','TN_16','TN_17','TN_18','TN_19','TN_20','TN_21','TN_22','TN_23','TN_24','TN_25','TN_26','TN_27','TN_28','TN_29',
         'TN_30','TN_31','TN_32','P_MSF','P_MF','P_DF','P_MSL','P_ML','P_DL',)
print('Number of columns to Drop: ',(len(columns)),end='\n')

data_drop=rfn.drop_fields(newdata,columns)
print('Number of Final columns: ',len(data_drop[0]),end='\n')
print('Column Names:\n',data_drop.dtype.names)
print(data_drop[:5])
```
